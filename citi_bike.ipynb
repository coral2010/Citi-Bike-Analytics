```python
# import dependencies
import pandas as pd
import datetime
```

```python
# read csv file
jan_2018 = pd.read_csv('Data/JC-201801-citibike-tripdata.csv', parse_dates=[0], infer_datetime_format=True)
feb_2018 = pd.read_csv('Data/JC-201802-citibike-tripdata.csv', parse_dates=[0], infer_datetime_format=True)
march_2018 = pd.read_csv('Data/JC-201803-citibike-tripdata.csv', parse_dates=[0], infer_datetime_format=True)
april_2018 = pd.read_csv('Data/JC-201804-citibike-tripdata.csv', parse_dates=[0], infer_datetime_format=True)
may_2018 = pd.read_csv('Data/JC-201805-citibike-tripdata.csv', parse_dates=[0], infer_datetime_format=True)
june_2018 = pd.read_csv('Data/JC-201806-citibike-tripdata.csv', parse_dates=[0], infer_datetime_format=True)
```

```python
jan_2018.head()
```

```python
tripdata_1h18 = jan_2018.append(feb_2018).append(march_2018).append(april_2018).append(may_2018).append(june_2018)
tripdata_1h18.head()
```

```python
# parse starttime and stoptime columns into separate date/time columns
temp = pd.DatetimeIndex(tripdata_1h18['starttime'])
tripdata_1h18['startDate'] = temp.date
tripdata_1h18['startTime'] = temp.time
del tripdata_1h18['starttime']

temp = pd.DatetimeIndex(tripdata_1h18['stoptime'])
tripdata_1h18['stopDate'] = temp.date
tripdata_1h18['stopTime'] = temp.time
del tripdata_1h18['stoptime']


tripdata_1h18.head()
```

```python
tripdata_1h18['gender'] = tripdata_1h18['gender'].replace({0:'unknown',
                                           1:'male',
                                           2:'female'})
```

```python
tripdata_1h18.head(20)
```

```python
tripdata_1h18.to_csv('data/tripdata_1h18.csv', header=True)
```
